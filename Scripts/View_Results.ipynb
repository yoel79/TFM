# Comparación de Métricas de Error por Modelo y Empresa

## Carga de Datos

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Estilo de las gráficas
plt.style.use('seaborn-v0_8')

# Empresas analizadas
empresas = ["AAPL", "AMZN", "TSLA", "GOOGL", "MSFT"]

# Datos actualizados de las métricas por modelo y empresa
mae_data = {
    "Empresa": empresas * 3,
    "Modelo": ["LSTM"] * 5 + ["Informer"] * 5 + ["TFT"] * 5,
    "MAE": [0.378, 0.080, 0.905, 0.235, 0.156,
            0.165, 0.068, 0.627, 0.231, 0.133,
            0.191, 0.106, 0.742, 0.215, 0.108]
}

rmse_data = {
    "Empresa": empresas * 3,
    "Modelo": ["LSTM"] * 5 + ["Informer"] * 5 + ["TFT"] * 5,
    "RMSE": [0.418, 0.103, 1.093, 0.265, 0.184,
             0.230, 0.092, 0.772, 0.273, 0.170,
             0.237, 0.133, 0.910, 0.263, 0.140]
}

mape_data = {
    "Empresa": empresas * 3,
    "Modelo": ["LSTM"] * 5 + ["Informer"] * 5 + ["TFT"] * 5,
    "MAPE": [18.850, 10.899, 142.711, 14.882, 9.391,
             8.190, 9.106, 128.630, 13.972, 8.259,
             10.239, 13.888, 122.896, 14.337, 6.511]
}

# Crear DataFrames a partir de los diccionarios
mae_df = pd.DataFrame(mae_data)
rmse_df = pd.DataFrame(rmse_data)
mape_df = pd.DataFrame(mape_data)

# (Opcional) Mostrar los DataFrames
print("MAE:")
print(mae_df.pivot(index='Empresa', columns='Modelo', values='MAE').round(3))
print("\nRMSE:")
print(rmse_df.pivot(index='Empresa', columns='Modelo', values='RMSE').round(3))
print("\nMAPE:")
print(mape_df.pivot(index='Empresa', columns='Modelo', values='MAPE').round(2))
```

## Mostrar Gráficos

```python
# Configuración de gráficos con Seaborn
fig, axes = plt.subplots(3, 1, figsize=(14, 14))

# Gráfico de MAE
sns.barplot(x="Empresa", y="MAE", hue="Modelo", data=mae_df, ax=axes[0], palette="mako")
axes[0].set_title("Comparación de MAE por Modelo y Empresa", fontsize=16, fontweight="bold")
axes[0].set_ylabel("MAE", fontsize=14, fontweight="bold")

# Gráfico de RMSE
sns.barplot(x="Empresa", y="RMSE", hue="Modelo", data=rmse_df, ax=axes[1], palette="rocket")
axes[1].set_title("Comparación de RMSE por Modelo y Empresa", fontsize=16, fontweight="bold")
axes[1].set_ylabel("RMSE", fontsize=14, fontweight="bold")

# Gráfico de MAPE
sns.barplot(x="Empresa", y="MAPE", hue="Modelo", data=mape_df, ax=axes[2], palette="YlOrBr")
axes[2].set_title("Comparación de MAPE por Modelo y Empresa", fontsize=16, fontweight="bold")
axes[2].set_ylabel("MAPE (%)", fontsize=14, fontweight="bold")

# Ajustar diseño
plt.tight_layout()
plt.show()
```
